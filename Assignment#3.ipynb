g) Add a new column that categorizes employees based on their performance rating:
->Excellent for ratings >= 4.5
->Good for ratings >= 4.0 but < 4.5
->Average for ratings < 4.0

```python
def rating(Rating):
    if Rating >= 4.5:
        return 'Excellent'
    elif Rating >= 4.0 and Rating<4.5:
        return 'Good'
    else:
        return 'Average'

df['Remark']=df['Rating'].apply(rating)
df
```
