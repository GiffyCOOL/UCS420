```python
import numpy as np
import matplotlib.pyplot as plt

A = float(input("Enter the value of A: "))
x_vals = np.linspace(-10, 10, 400)
y_parabola = A * (x_vals**2)
y_sine = A * np.sin(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_parabola, label=r'$y = Ax^2$', color='blue', linestyle='--')
plt.plot(x_vals, y_sine, label=r'$y = A\sin(x)$', color='red', linestyle='-')

plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Graphs of y = Ax^2 and y = Asin(x)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
data_frame = pd.DataFrame({'Subject': ['Math', 'Physics', 'Chemistry', 'Biology', 'English'],'Scores': [99, 80, 78, 58, 62]})
plt.figure(figsize=(8, 5))
axis = sb.barplot(x='Subject', y='Scores', hue='Subject', data=data_frame, palette='husl', legend=False)
for bar in axis.patches:axis.annotate(f'{bar.get_height()}', (bar.get_x() + bar.get_width() / 2, bar.get_height()),ha='center', va='bottom', fontsize=12, fontweight='bold')
plt.title("Scores in Different Subjects", fontsize=14, fontweight='bold')
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
seed_value = 102317289
np.random.seed(seed_value)
random_data = np.random.randn(50)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].plot(np.cumsum(random_data), linestyle='-', color='black')
axes[0, 0].set_title("Cumulative Sum Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True)
noise_data = np.random.randn(50) * 0.5
axes[0, 1].scatter(range(50), random_data + noise_data, color='black', alpha=0.7)
axes[0, 1].set_title("Scatter Plot with Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Value")
axes[0, 1].grid(True)
axes[1, 0].axis("off")
axes[1, 1].axis("off")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data_source = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
data_table = pd.read_csv(data_source)
plt.figure(figsize=(10, 5))
sns.lineplot(x=data_table["month_number"], y=data_table["total_profit"], marker="o", label="Total Profit", color="blue")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("1) Total Profit Per Month")
plt.legend()
plt.grid(True)
plt.show()
plt.figure(figsize=(12, 6))
for field in data_table.columns[1:-1]:  # Excluding 'month_number' and 'total_profit'
    sns.lineplot(x=data_table["month_number"], y=data_table[field], marker="o", label=field)
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.title("2) Sales of Different Products Per Month")
plt.legend()
plt.grid(True)
plt.show()
data_table.set_index("month_number").plot(kind="bar", figsize=(15, 6))
plt.xlabel("Month Number")
plt.ylabel("Values")
plt.title("3) Bar Chart of All Features")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.grid(axis="y", linestyle="--")
plt.show()
```
