```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Part I: Randomized Sales Data Generation (NumPy)
seed_value = 102317288  
np.random.seed(seed_value)
# Generate Sales Data
months_list = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
categories = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]
sales_data = np.random.randint(1000, 5000, size=(12, 4))
# Convert to DataFrame
sales_df = pd.DataFrame(sales_data, columns=categories, index=months_list)
# Part II: Data Manipulation and Analysis (Pandas)
# Display first 5 rows and summary statistics
print("First 5 Rows:\n", sales_df.head())
print("\nSummary Statistics:\n", sales_df.describe())
# Calculate total sales per category and per month
total_sales_per_category = sales_df.sum()
total_sales_per_month = sales_df.sum(axis=1)
sales_df["Total Sales"] = total_sales_per_month
# Calculate average sales growth between consecutive months
sales_df["Growth Rate"] = sales_df["Total Sales"].pct_change() * 100
# Apply discount based on roll number
if seed_value % 2 == 0:
    sales_df["Electronics"] *= 0.90  # 10% discount on Electronics
else:
    sales_df["Clothing"] *= 0.85  # 15% discount on Clothing
# Part III: Visualizations (Matplotlib and Seaborn)
# Plot monthly sales trends
plt.figure(figsize=(10, 5))
for category in categories:
    plt.plot(sales_df.index, sales_df[category], marker='o', label=category)
plt.xlabel("Month")
plt.ylabel("Sales (units)")
plt.title("Monthly Sales Trends by Category")
plt.legend()
plt.grid()
plt.show()
# Box plots to show sales distribution
plt.figure(figsize=(8, 5))
sns.boxplot(data=sales_df[categories])
plt.title("Sales Distribution by Category")
plt.ylabel("Sales (units)")
plt.show()
# Printing final DataFrame
sales_df
```

```python
import numpy as np
matrix = np.array([[1, -2, 3], [-4, 5, -6]])
absolute_matrix = np.abs(matrix)
print("(i) Element-wise Absolute Value:\n", absolute_matrix)
flattened_percentiles = np.percentile(matrix, [25, 50, 75])  # For the flattened array
column_percentiles = np.percentile(matrix, [25, 50, 75], axis=0)  # For each column
row_percentiles = np.percentile(matrix, [25, 50, 75], axis=1)  # For each row
print("(ii)")
print("\tPercentiles (Flattened):", flattened_percentiles)
print("\tPercentiles (Columns):\n", column_percentiles)
print("\tPercentiles (Rows):\n", row_percentiles)
mean_flat = np.mean(matrix)
median_flat = np.median(matrix)
std_flat = np.std(matrix)
mean_col = np.mean(matrix, axis=0)
median_col = np.median(matrix, axis=0)
std_col = np.std(matrix, axis=0)
mean_row = np.mean(matrix, axis=1)
median_row = np.median(matrix, axis=1)
std_row = np.std(matrix, axis=1)
print("(iii)")
print("\tFlattened Mean:", mean_flat)
print("\tFlattened Median:", median_flat)
print("\tFlattened Standard Deviation:", std_flat)
print("\tColumn-wise Mean:", mean_col)
print("\tColumn-wise Median:", median_col)
print("\tColumn-wise Standard Deviation:", std_col)
print("\tRow-wise Mean:", mean_row)
print("\tRow-wise Median:", median_row)
print("\tRow-wise Standard Deviation:", std_row)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data_source = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
data_table = pd.read_csv(data_source)
plt.figure(figsize=(10, 5))
sns.lineplot(x=data_table["month_number"], y=data_table["total_profit"], marker="o", label="Total Profit", color="blue")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("1) Total Profit Per Month")
plt.legend()
plt.grid(True)
plt.show()
plt.figure(figsize=(12, 6))
for field in data_table.columns[1:-1]:  # Excluding 'month_number' and 'total_profit'
    sns.lineplot(x=data_table["month_number"], y=data_table[field], marker="o", label=field)
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.title("2) Sales of Different Products Per Month")
plt.legend()
plt.grid(True)
plt.show()
data_table.set_index("month_number").plot(kind="bar", figsize=(15, 6))
plt.xlabel("Month Number")
plt.ylabel("Values")
plt.title("3) Bar Chart of All Features")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.grid(axis="y", linestyle="--")
plt.show()
```
