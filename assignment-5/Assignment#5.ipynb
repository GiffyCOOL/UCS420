```python
import numpy as np
import matplotlib.pyplot as plt

# Q1: Sum operations on matrix
gfg = np.array([[4, 1, 9], [12, 3, 1], [4, 5, 6]])
print("Sum of all elements:", np.sum(gfg))
print("Row-wise sum:", np.sum(gfg, axis=1))
print("Column-wise sum:", np.sum(gfg, axis=0))
```

```python
# Q2a: Sorting and finding elements
array = np.array([10, 52, 62, 16, 16, 54, 453])
sorted_array = np.sort(array)
indices_sorted = np.argsort(array)
print("Sorted array:", sorted_array)
print("Indices of sorted array:", indices_sorted)
print("4 smallest elements:", sorted_array[:4])
print("5 largest elements:", sorted_array[-5:])
```

```python
# Q2b: Integer and float elements
array = np.array([10, 52, 62, 16, 16, 54, 453])
print(f"(a) For array {array} :\n")
print("  (i) Sorted array:", np.sort(array))
print("  (ii) Indices of sorted array:", np.argsort(array))
print("  (iii) 4 smallest elements:", np.sort(array)[:4])
print("  (iv) 5 largest elements:", np.sort(array)[-5:])
print();

array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
print(f"(b) For array {array} :\n")
print("  (i) Integer elements only:",array[array == np.floor(array)])
print("  (ii) Float elements only:", array[array != np.floor(array)])
```

```python
# Q3: Sales dataset operations
X = 148  # ASCII sum of A and S  # Replace "AB" with your initials
sales = np.array([X, X+50, X+100, X+150, X+200])
tax_rate = ((X % 5) + 5) / 100
sales_after_tax = sales * (1 - tax_rate)

sales_discounted = np.where(sales < X+100, sales * 0.95, sales * 0.90)
sales_weeks = np.vstack([sales] * 3)
sales_weeks = sales_weeks * np.array([1.00, 1.02, 1.04]).reshape(3, 1)

print("(b) Computing personalized tax rate ((X % 5) + 5) / 100 :")
print("  (i) Tax rate:", tax_rate)
print("  (ii) Sales after tax:", sales_after_tax)
print("(c) Discounted Sales:", sales_discounted)
print("(d) Weekly Sales after Increase:", sales_weeks)
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-10, 10, 100)


y1 = x**2                      
y2 = np.sin(x)                 
y3 = np.exp(x)                 
y4 = np.log(np.abs(x) + 1)     


plt.figure(figsize=(12, 8))

# GRAPH OF Y = x^2
plt.subplot(2, 2, 1)
plt.plot(x, y1, label=r"$Y = x^2$", color='r')
plt.title("Quadratic Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()

# GRAPH OF Y = sin(x)
plt.subplot(2, 2, 2)
plt.plot(x, y2, label=r"$Y = \sin(x)$", color='g')
plt.title("Sine Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()

# GRAPH OF Y = e^x
plt.subplot(2, 2, 3)
plt.plot(x, y3, label=r"$Y = e^x$", color='b')
plt.title("Exponential Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()

# GRAPH OF Y = log(|x| + 1)
plt.subplot(2, 2, 4)
plt.plot(x, y4, label=r"$Y = \log(|x| + 1)$", color='purple')
plt.title("Logarithmic Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
